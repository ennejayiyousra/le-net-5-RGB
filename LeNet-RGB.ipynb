```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from PIL import Image
from collections import defaultdict
import cv2

class LeNet5Model:
    """
    LeNet-5 CNN Architecture modified for Tifinagh character classification (33 classes)
    
    Architecture:
    Input (32x32x1) -> C1 (Conv 5x5, 6 filters) -> S2 (AvgPool 2x2) -> 
    C3 (Conv 5x5, 16 filters) -> S4 (AvgPool 2x2) -> C5 (FC 120) -> 
    F6 (FC 84) -> Output (FC 33)
    """
    
    def __init__(self, num_classes=33):
        self.num_classes = num_classes
        self.model = self._build_model()
        
    def _build_model(self):
        """Build the LeNet-5 model for RGB images"""
        model = models.Sequential([
            # Input layer - expecting 32x32x3 RGB images
            layers.Input(shape=(32, 32, 3)),
            
            # C1: Convolutional layer - 5x5 kernel, 6 filters, RGB input
            layers.Conv2D(filters=6, kernel_size=(5, 5), strides=(1, 1), 
                         padding='valid', activation='tanh', name='c1'),
            
            # S2: Average Pooling layer - 2x2 kernel
            layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2), name='s2'),
            
            # C3: Convolutional layer - 5x5 kernel, 16 filters
            layers.Conv2D(filters=16, kernel_size=(5, 5), strides=(1, 1), 
                         padding='valid', activation='tanh', name='c3'),
            
            # S4: Average Pooling layer - 2x2 kernel
            layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2), name='s4'),
            
            # Flatten for fully connected layers
            layers.Flatten(),
            
            # C5: Fully connected layer - 120 neurons
            layers.Dense(120, activation='tanh', name='c5'),
            
            # F6: Fully connected layer - 84 neurons
            layers.Dense(84, activation='tanh', name='f6'),
            
            # Output: Classification layer
            layers.Dense(self.num_classes, activation='softmax', name='output')
        ])
        
        return model
    
    def get_model(self):
        """Return the model"""
        return self.model
    
    def summary(self):
        """Print model summary"""
        return self.model.summary()
    
    def get_feature_extractor(self, layer_name):
        """Create a model to extract features from a specific layer"""
        try:
            layer_output = self.model.get_layer(layer_name).output
            feature_extractor = models.Model(inputs=self.model.input, outputs=layer_output)
            return feature_extractor
        except ValueError as e:
            print(f"Layer {layer_name} not found. Available layers:")
            for layer in self.model.layers:
                print(f"  - {layer.name}")
            return None

class TifinaghDataGenerator:
    """Custom data generator for Tifinagh characters from AMHCD_64 dataset"""
    
    def __init__(self, data_dir, batch_size=32, target_size=(32, 32), train_ratio=0.8):
        self.data_dir = data_dir
        self.batch_size = batch_size
        self.target_size = target_size
        self.train_ratio = train_ratio
        self.class_names = []
        self.class_to_idx = {}
        self.idx_to_class = {}
        
        self._prepare_data_structure()
        
    def _prepare_data_structure(self):
        """Prepare data structure and class mappings"""
        # Get all class directories
        all_items = os.listdir(self.data_dir)
        self.class_names = sorted([item for item in all_items 
                                  if os.path.isdir(os.path.join(self.data_dir, item))])
        
        print(f"Found {len(self.class_names)} classes: {self.class_names}")
        
        self.class_to_idx = {cls_name: idx for idx, cls_name in enumerate(self.class_names)}
        self.idx_to_class = {idx: cls_name for cls_name, idx in self.class_to_idx.items()}
        
        # Print class distribution
        for class_name in self.class_names:
            class_dir = os.path.join(self.data_dir, class_name)
            if os.path.isdir(class_dir):
                num_images = len([f for f in os.listdir(class_dir) 
                                if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))])
                train_size = int(num_images * self.train_ratio)
                val_size = num_images - train_size
                print(f"Class '{class_name}': {num_images} total, {train_size} train, {val_size} val")
    
    def create_data_generators(self):
        """Create training and validation data generators for RGB images"""
        # Data augmentation for training - RGB images
        train_datagen = ImageDataGenerator(
            rescale=1./255,  # Scale to [0, 1]
            rotation_range=10,
            width_shift_range=0.1,
            height_shift_range=0.1,
            horizontal_flip=False,  # Usually not good for text/characters
            validation_split=1.0-self.train_ratio  # Reserve portion for validation
        )
        
        # Only rescaling for validation
        val_datagen = ImageDataGenerator(
            rescale=1./255,  # Scale to [0, 1]
            validation_split=1.0-self.train_ratio
        )
        
        # Training generator
        train_generator = train_datagen.flow_from_directory(
            self.data_dir,
            target_size=self.target_size,
            batch_size=self.batch_size,
            color_mode='rgb',  # RGB images
            class_mode='categorical',
            subset='training',
            shuffle=True,
            seed=42
        )
        
        # Validation generator
        val_generator = val_datagen.flow_from_directory(
            self.data_dir,
            target_size=self.target_size,
            batch_size=self.batch_size,
            color_mode='rgb',  # RGB images
            class_mode='categorical',
            subset='validation',
            shuffle=False,
            seed=42
        )
        
        return train_generator, val_generator

class LeNet5Trainer:
    """Training and evaluation class for LeNet-5 in TensorFlow"""
    
    def __init__(self, model, class_names=None):
        self.model = model
        self.class_names = class_names
        self.history = None
        
    def compile_model(self, optimizer='adam', learning_rate=0.001):
        """Compile the model with optimizer and loss function"""
        if optimizer.lower() == 'sgd':
            opt = optimizers.SGD(learning_rate=learning_rate, momentum=0.9)
        elif optimizer.lower() == 'adam':
            opt = optimizers.Adam(learning_rate=learning_rate)
        elif optimizer.lower() == 'rmsprop':
            opt = optimizers.RMSprop(learning_rate=learning_rate)
        elif optimizer.lower() == 'adagrad':
            opt = optimizers.Adagrad(learning_rate=learning_rate)
        else:
            opt = optimizers.Adam(learning_rate=learning_rate)
            
        self.model.compile(
            optimizer=opt,
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        print(f"Model compiled with {optimizer} optimizer (lr={learning_rate})")
    
    def train(self, train_generator, val_generator, epochs=50, verbose=1):
        """Train the model"""
        print(f"Starting training for {epochs} epochs...")
        print(f"Training samples: {train_generator.samples}")
        print(f"Validation samples: {val_generator.samples}")
        
        # Calculate steps per epoch
        steps_per_epoch = train_generator.samples // train_generator.batch_size
        validation_steps = val_generator.samples // val_generator.batch_size
        
        # Callbacks
        callbacks = [
            keras.callbacks.EarlyStopping(
                monitor='val_accuracy',
                patience=10,
                restore_best_weights=True
            ),
            keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=5,
                min_lr=1e-7
            )
        ]
        
        # Train the model
        self.history = self.model.fit(
            train_generator,
            steps_per_epoch=steps_per_epoch,
            epochs=epochs,
            validation_data=val_generator,
            validation_steps=validation_steps,
            callbacks=callbacks,
            verbose=verbose
        )
        
        print("Training completed!")
        
    def plot_training_curves(self):
        """Plot training and validation curves"""
        if self.history is None:
            print("No training history found. Train the model first.")
            return
            
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
        
        # Loss curves
        epochs = range(1, len(self.history.history['loss']) + 1)
        ax1.plot(epochs, self.history.history['loss'], 'b-', label='Training Loss')
        ax1.plot(epochs, self.history.history['val_loss'], 'r-', label='Validation Loss')
        ax1.set_title('Training and Validation Loss')
        ax1.set_xlabel('Epochs')
        ax1.set_ylabel('Loss')
        ax1.legend()
        ax1.grid(True)
        
        # Accuracy curves
        ax2.plot(epochs, self.history.history['accuracy'], 'b-', label='Training Accuracy')
        ax2.plot(epochs, self.history.history['val_accuracy'], 'r-', label='Validation Accuracy')
        ax2.set_title('Training and Validation Accuracy')
        ax2.set_xlabel('Epochs')
        ax2.set_ylabel('Accuracy')
        ax2.legend()
        ax2.grid(True)
        
        plt.tight_layout()
        plt.show()
    
    def evaluate_model(self, val_generator):
        """Evaluate model and create confusion matrix"""
        print("Evaluating model...")
        
        # Get predictions
        val_generator.reset()
        predictions = self.model.predict(val_generator, verbose=1)
        predicted_classes = np.argmax(predictions, axis=1)
        
        # Get true labels
        true_classes = val_generator.classes
        
        # Create confusion matrix
        cm = confusion_matrix(true_classes, predicted_classes)
        
        # Plot confusion matrix
        plt.figure(figsize=(12, 10))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.class_names if self.class_names else range(len(cm)),
                    yticklabels=self.class_names if self.class_names else range(len(cm)))
        plt.title('Confusion Matrix - Tifinagh Character Classification')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        # Print classification report
        if self.class_names:
            print("\nClassification Report:")
            print(classification_report(true_classes, predicted_classes, 
                                      target_names=self.class_names))
        
        # Calculate and print accuracy
        accuracy = np.mean(predicted_classes == true_classes)
        print(f"\nFinal Accuracy: {accuracy:.4f}")
        
        return accuracy, cm
    
    def visualize_feature_maps(self, sample_image, layer_name='c1'):
        """Visualize feature maps from a specific layer for RGB input"""
        # Create feature extractor
        lenet_model = LeNet5Model(num_classes=len(self.class_names) if self.class_names else 33)
        lenet_model.model = self.model
        feature_extractor = lenet_model.get_feature_extractor(layer_name)
        
        if feature_extractor is None:
            return
        
        # Ensure proper input shape for RGB image
        if len(sample_image.shape) == 3 and sample_image.shape[-1] == 3:  # RGB image
            sample_image = np.expand_dims(sample_image, axis=0)   # Add batch dim
        elif len(sample_image.shape) == 2:  # If somehow 2D, expand to RGB
            sample_image = np.stack([sample_image] * 3, axis=-1)  # Convert to RGB
            sample_image = np.expand_dims(sample_image, axis=0)   # Add batch dim
        
        print(f"Input shape for feature extraction: {sample_image.shape}")
        
        # Get feature maps
        features = feature_extractor.predict(sample_image)
        
        # Remove batch dimension if present
        if len(features.shape) == 4:
            features = features[0]
        
        print(f"Feature maps shape: {features.shape}")
        
        # Plot feature maps
        if len(features.shape) == 3:  # Conv layer output (H, W, C)
            num_features = features.shape[-1]
            
            # Calculate grid size
            cols = int(np.ceil(np.sqrt(num_features)))
            rows = int(np.ceil(num_features / cols))
            
            fig, axes = plt.subplots(rows, cols, figsize=(15, 15))
            fig.suptitle(f'Feature Maps from Layer {layer_name.upper()} - RGB Input', fontsize=16)
            
            for i in range(rows * cols):
                if rows == 1:
                    ax = axes[i] if cols > 1 else axes
                else:
                    row, col = i // cols, i % cols
                    ax = axes[row, col]
                
                if i < num_features:
                    feature_map = features[:, :, i]
                    im = ax.imshow(feature_map, cmap='viridis')
                    ax.set_title(f'Filter {i+1}', fontsize=10)
                    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
                else:
                    ax.axis('off')
                
                ax.set_xticks([])
                ax.set_yticks([])
            
            plt.tight_layout()
            plt.show()
        else:
            print(f"Layer {layer_name} output shape: {features.shape}")
            print("This layer doesn't produce 2D feature maps for visualization")

def create_sample_predictions(model, val_generator, class_names, num_samples=9):
    """Create a visualization of sample predictions for RGB images"""
    val_generator.reset()
    
    # Get a batch of validation images
    sample_batch = next(val_generator)
    images, true_labels = sample_batch
    
    # Make predictions
    predictions = model.predict(images[:num_samples])
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(true_labels[:num_samples], axis=1)
    
    # Plot samples with predictions
    fig, axes = plt.subplots(3, 3, figsize=(12, 12))
    fig.suptitle('Sample Predictions - RGB Images', fontsize=16)
    
    for i, ax in enumerate(axes.flat):
        if i < num_samples:
            # Display RGB image
            img = images[i]
            # Ensure values are in [0, 1] range for display
            img_display = np.clip(img, 0, 1)
            ax.imshow(img_display)
            
            # Add prediction info
            true_class = class_names[true_classes[i]] if class_names else true_classes[i]
            pred_class = class_names[predicted_classes[i]] if class_names else predicted_classes[i]
            confidence = predictions[i][predicted_classes[i]]
            
            color = 'green' if predicted_classes[i] == true_classes[i] else 'red'
            ax.set_title(f'True: {true_class}\nPred: {pred_class}\nConf: {confidence:.3f}', 
                        color=color, fontsize=10)
            
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()

def mathematical_formulas():
    """
    Mathematical formulas used in LeNet-5:
    
    1. Convolution Operation:
       Y[i,j] = Σ Σ X[i+m, j+n] * W[m,n] + b
       
    2. Average Pooling:
       Y[i,j] = (1/k²) * Σ Σ X[i*s+m, j*s+n]
       where k is kernel size, s is stride
       
    3. Tanh Activation:
       f(x) = tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))
       
    4. Softmax (output layer):
       σ(z_i) = e^(z_i) / Σ e^(z_j) for j=1 to K
       
    5. Categorical Cross-Entropy Loss:
       L = -Σ y_i * log(ŷ_i)
       where y_i is true label, ŷ_i is predicted probability
       
    6. Adam Optimizer:
       m_t = β₁ * m_{t-1} + (1-β₁) * ∇L(θ_t)
       v_t = β₂ * v_{t-1} + (1-β₂) * (∇L(θ_t))²
       θ_{t+1} = θ_t - η * m_t / (√v_t + ε)
    """
    pass

def main():
    """Main training function for AMHCD_64 dataset"""
    # Set random seeds for reproducibility
    tf.random.set_seed(42)
    np.random.seed(42)
    
    # Check GPU availability
    print("GPU Available: ", tf.config.list_physical_devices('GPU'))
    print("TensorFlow Version: ", tf.__version__)
    
    # Dataset path (update this to your actual path)
    data_dir = r"C:\Users\yousra\Desktop\University\lenet\dataset\AMHCD_64\AMHCD_64"
    # Alternative path format: "./dataset/AMHCD_64/AMHCD_64"
    
    # Hyperparameters
    batch_size = 32
    learning_rate = 0.001
    epochs = 10
    target_size = (32, 32)  # LeNet-5 input size
    
    # Check if dataset exists
    if not os.path.exists(data_dir):
        print(f"Dataset directory not found: {data_dir}")
        print("Please update the data_dir path in the main() function")
        return None
    
    try:
        # Create data generators
        print("Preparing data generators...")
        data_gen = TifinaghDataGenerator(data_dir, batch_size=batch_size, 
                                       target_size=target_size, train_ratio=0.8)
        train_generator, val_generator = data_gen.create_data_generators()
        
        # Get number of classes
        num_classes = len(data_gen.class_names)
        print(f"Detected {num_classes} classes in dataset")
        
        # Create model
        print("Creating LeNet-5 model...")
        lenet_model = LeNet5Model(num_classes=num_classes)
        model = lenet_model.get_model()
        
        # Print model summary
        print("\nModel Architecture:")
        model.summary()
        
        # Create trainer
        trainer = LeNet5Trainer(model, data_gen.class_names)
        
        # Compile model
        trainer.compile_model(optimizer='adam', learning_rate=learning_rate)
        
        print(f"\n{'='*50}")
        print(f"Training LeNet-5 on Tifinagh AMHCD_64 Dataset")
        print(f"{'='*50}")
        
        # Train the model
        trainer.train(train_generator, val_generator, epochs=epochs)
        
        # Plot training curves
        trainer.plot_training_curves()
        
        # Evaluate model
        accuracy, confusion_mat = trainer.evaluate_model(val_generator)
        
        # Create sample predictions
        create_sample_predictions(model, val_generator, data_gen.class_names)
        
        # Visualize feature maps
        val_generator.reset()
        sample_batch = next(val_generator)
        sample_image = sample_batch[0][0]  # First image from batch
        
        trainer.visualize_feature_maps(sample_image, 'c1')
        trainer.visualize_feature_maps(sample_image, 'c3')
        
        # Save the trained model
        model.save('lenet5_tifinagh_model.h5')
        
        # Save additional information
        model_info = {
            'class_names': data_gen.class_names,
            'class_to_idx': data_gen.class_to_idx,
            'num_classes': num_classes,
            'target_size': target_size,
            'final_accuracy': float(accuracy)
        }
        
        import json
        with open('model_info.json', 'w') as f:
            json.dump(model_info, f)
        
        print(f"\nModel saved as 'lenet5_tifinagh_model.h5'")
        print(f"Model info saved as 'model_info.json'")
        
        return model, trainer, data_gen
        
    except Exception as e:
        print(f"Error during training: {e}")
        import traceback
        traceback.print_exc()
        return None

def test_model_architecture():
    """Test the model architecture with dummy RGB data"""
    print("Testing LeNet-5 architecture for RGB images...")
    
    # Create model
    lenet_model = LeNet5Model(num_classes=33)
    model = lenet_model.get_model()
    
    # Print model summary
    print("\nModel Architecture:")
    model.summary()
    
    # Test with dummy RGB input
    dummy_input = tf.random.normal([1, 32, 32, 3])  # RGB input
    
    try:
        output = model(dummy_input)
        print(f"\nArchitecture Test Results:")
        print(f"Input shape: {dummy_input.shape} (RGB)")
        print(f"Output shape: {output.shape}")
        print(f"Output type: {type(output)}")
        
        # Test feature extraction
        feature_extractor = lenet_model.get_feature_extractor('c1')
        if feature_extractor:
            c1_features = feature_extractor(dummy_input)
            print(f"C1 features shape: {c1_features.shape}")
            
        print("✅ RGB Architecture test passed!")
        
    except Exception as e:
        print(f"❌ Architecture test failed: {e}")

def load_trained_model(model_path='lenet5_tifinagh_model.h5', info_path='model_info.json'):
    """Load a trained model and its information"""
    try:
        # Load model
        model = keras.models.load_model(model_path)
        print(f"Model loaded from {model_path}")
        
        # Load model info
        import json
        with open(info_path, 'r') as f:
            model_info = json.load(f)
        
        print(f"Model info loaded from {info_path}")
        print(f"Classes: {len(model_info['class_names'])}")
        print(f"Final training accuracy: {model_info['final_accuracy']:.4f}")
        
        return model, model_info
        
    except Exception as e:
        print(f"Error loading model: {e}")
        return None, None

if __name__ == "__main__":
    # Test the architecture first
    print("Testing model architecture...")
    test_model_architecture()
    
    print("\n" + "="*60)
    print("Starting training on AMHCD_64 dataset...")
    print("="*60)
    
    # Run main training
    main()
    
    if result is None:
        print("\nTraining failed. Please check your dataset path.")
        print("Expected structure:")
        print("dataset/AMHCD_64/AMHCD_64/")
        print("├── ya/")
        print("├── yab/")
        print("├── yad/")
        print("└── ... (other Tifinagh classes)")
    else:
        print("\nTraining completed successfully!")
        print("Check the generated plots for training curves, confusion matrix, and sample predictions.")
        
        # Example of how to load the model later
        print("\n" + "="*40)
        print("Testing model loading...")
        loaded_model, model_info = load_trained_model()
        if loaded_model:
            print("✅ Model loading test passed!")
        else:
            print("❌ Model loading test failed!")
```
